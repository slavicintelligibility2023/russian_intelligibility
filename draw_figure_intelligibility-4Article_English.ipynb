## INTRODUCTION
It has long been recognized that speakers of related languages can have successful communication speaking their native languages. Various terms have been used to refer to such kind of communication like semicommunication, receptive multilingualism, semibilingualism, inherent intelligibility, non-convergent discourse, asymmetric/bilingual discourse (Gooskens, 2013).

Mutual intelligibility has long been used as one of the criteria to identify whether two speech varieties should be considered dialects of the same language or two distinct languages (Gleason 1959, Comrie 1981). Knowledge about the degree to which two related languages are mutually intelligible is also required for various language planning, language policy, language contact issues.

Receptive multilingualism has recently come into focus as a way of communicating across language borders, a possible alternative to using a lingua franca. Two major projects dedicated to mutual intelligibility studies need to be mentioned here. 

The first one, MICReLa (Mutual Intelligibility of Closely Related Languages), focuses on receptive multilingualism in Europe with an aim to improve communication in the region while preserving linguistic diversity (Moberg et al., 2007; Gooskens, 2007; Gooskens & Swarte, 2017; Gooskens et al., 2018). The project studies what factors influence mutual intercomprehension (both spoken and written), as well as to what extent it can be used to overcome language barriers. It follows the team’s preceding research into intelligibility between Scandinavian languages and currently covers 16 languages spoken in Europe from Germanic, Romance, and Slavic language families. Though this large-scale study provides an important basis for comparison, the resulting picture of mutual intelligibility between languages of these families is incomplete. The Slavic language family in this project is represented by the West and South Slavic languages. As it will be shown below, the missing information on the East Slavic languages (and most importantly Russian) can help to understand the results for the South Slavic group better and see them in a slightly different light.

The second major project, INCOMSLAV (Mutual Intelligibility and Surprisal in Slavic Intercomprehension), aims to research the relation between information density, encoding density and grammaticalisation in a cross-linguistic perspective with a focus on mutual intelligibility between Slavic languages (Mosbach et al., 2019; Stenger et al., 2020). Within this framework spoken and written intercomprehension experiments between 11 Slavic languages (including Russian) are conducted through a web-based platform. At the current stage the project includes testing comprehension on three levels: individual words, word combinations, and sentences.

Our paper presents a study of written intelligibility of eight Slavic languages to native speakers of Russian. It aims to check how intelligible words of an unknown but related language can be when given in a text. 


## Experiment set-up

In our experiment participants were given a text in Russian with omitted words and a parallel text in a Slavic language (unknown to them) to use as a clue. In our case, there were no restrictions on the answers other than the ones imposed by the context in Russian and the clue from a corresponding word in another Slavic language (provided the participant identified it correctly in the sentence). This way of arranging the experiment does not limit participants in their choice of words and allows gathering objective quantitative information on a text intelligibility. 

When choosing words to omit in the test, we selected different parts of speech except for function words (because of their ambiguous nature). 

If a word was omitted in the text once, it was to be omitted in all other occurrences in this test (so that it would not be possible to guess it in case its equivalent is repeated in the parallel text). Words (totalling 186 lemmas, 232 word occurrences) were omitted in the Russian part of a test independently from translations.

Of the nine languages used in the study, five (including Russian) use Cyrillic alphabet and four use Latin alphabet. To avoid problems with the extended Latin script, we transliterated such parallel texts into Cyrillic using simple rules of substitution. For the sake of consistency all parallel texts written in Cyrillic script were transliterated into standard Latin. Thus, subjects saw each parallel text written both in Latin and Cyrillic scripts aligned by sentences.

As it was mentioned in (Krutchenko et al., 2019), it is possible to restore words in a gapped text using the text redundancy on the basis of limitations imposed by the context itself. To identify this influence each of the six test pieces was given to a control group of Russian native speakers to do without a parallel text in another Slavic language.

All participants did tests on a web site without any observation of their behaviour. At the beginning each participant was asked to select their level of education, to indicate if they knew any other Slavic language, and if their education included any language studies. All Slavic languages named by the participant were excluded from the list from which the parallel language for the test was randomly selected. It was also possible to skip these three questions, then the subject was given a control test (a gapped text in Russian without a parallel text).

In the experiment we used three pieces from M. Bulgakov’s Master and Margaret (originally written in Russian), three pieces of H. Sienkiewicz’s Quo Vadis? (originally written in Polish), and their modern translations into the other languages in the study.

### Description of Data Structure

The following files are placed to the *data* folder. The first number in every file is the row id.

- **user_data.csv** - participants' data. Fields description:
    - *id_user* - identifier of a participant.
    - *test_id* - identifier of a test passed by the participant.
    - *parallel_lang* - a language used for a parallel text. *No Parallel Text* denotes the control group, i.e. those participants who restored words without the parallel text. 
    - *age* - age group of the participant. *sch* denotes school children, bak - bachelor students, mag - master student, fin - finished his or her education.
    - *speciality* - the participant's education included any language studies (translation, linguistics, philology): yes - includes, no - don't includes or was not selected.
    - *known_langs* - if the participant knows any other Slavic language. Five and more languages denote situation when the participant declined any parallel text and volonteered to be in the control group.
    - *ans_cnt* - number of non-empty answers for this participant.
    
    
- **user_answers.csv** - participants' answers. Fields description:
    - *id_user* - identifier of a participant; equals to the same field in the *user_data* table.
    - *test_id* - identifier of a test passed by the participant.
    - *parallel_lang* - a language used for a parallel text; equals to the same field in the *user_data* table.
    - *answer_no* - position of a omitted word in the test.
    - *answer* - participant's answer.
    - *correctness* - asessors' evaluation of the answer. incorr denotes an incorrect answer, pcorr - partially correct anser, corr - correct answer.
    
    
- **texts.csv** - text phragments used as tests. The words, marked by the underscore denote omitted words in parallel texts. All texts are aligned by sentences, i.e. have the same identifier for different languages. Fields description:
    - *id_set* - identifier of a text.
    - *lang_name* - language of the text.
    - *sent_pos* - position of a sentence in the text.
    - *sent* - text of the sentence.

## INTELLIGIBILITY OF WORD CLASSES AND INTELLIGIBILITY OF A TEST

Our second hypothesis was that the intelligibility of an unknown foreign text depends not only on the percentage of cognate words (or non-cognate words alternatively) in this text, but also on the percentage of cognate synonyms and false friends as discussed above. 

Thus we specifically focused on the type of relation between the omitted Russian word and its equivalent in the parallel text in an attempt to evaluate its influence on the general intelligibility of the parallel text.
To do it we classified all pairs of omitted words and their equivalents in our experiment into the following four groups. 
The first class consisted of such words from parallel texts which were perceived as derived from the same root as the original omitted word, e.g.,вырвался \[‘vɨrvəɫs⁽ʲ⁾ə\] (Rus), вырваўся \[‘vɨrvawsʲa\] (Bel), wyrwał (Pol) ‘broke free’. (All examples here and below are taken from our parallel texts.)
The second class consisted of parallel text words which were perceived as cognate synonyms, i.e. having synonyms or semantically close words that are regarded as cognates to the omitted one,  e.g. дорогу \[dɐˈroɡʊ\] (Rus), пътя \[pɤ‘tʲɐ\] (Bulg), пут [pûːt] (Serb) ‘road’. Here the Bulgarian and the Serbian words both are perceived as similar to the Russian word путь [putʲ] ‘way’’ which is a synonym to the word дорога [dɐˈroɡə]. 
The third class consisted of non-cognates, i.e. totally unrelated words having no written or phonetic correspondence with the omitted one, e.g. трубою [trʊˈbojʊ] (Rus), сурмою [sur’mojʊ] (Ukr), poľnicou (Slovak) ‘trumpet’. 
The fourth class consisted of false friends to the omitted Russian word, the ones that are perceived as similar but have different translations (Frunza, 2006) e.g. спиною [spʲɪˈnojʊ] (Rus), плячыма [plʲaˈtʂɨma] (Bel), гърба [‘ɡɤrba] (Bulg) ‘human back’. Here the Belarusian word плячыма [plʲaˈtʂɨma] may be associated with the Russian word плечами [plʲɪˈt͡ɕæmʲɪ], an instrumental form of the Russian word плечи [ˈplʲet͡ɕɪ] ‘shoulders’. The Bulgarian word гърба [‘ɡɤrba] can be associated with the Russian word горба [ɡɐrˈba], a genitive form of the word горб [ɡorp] ‘humpback’. 

0 - (non-cognates) nothing similar, 1 - same word root, 2 - cognate synonyms (has word with a similar sence), ! - false translator's friend.

Now we use [Power Analysis](https://en.wikipedia.org/wiki/Power_(statistics)) \([or read here](http://www.statskingdom.com/doc_test_power.html)\). It's calculated as following ![](https://wikimedia.org/api/rest_v1/media/math/render/svg/d7326aa9d5d7c4c8bcbb9c569949cd1e53203a88). It should be >95%, but we are not substractin 1, so it should be <5%.

Note that the Power analysis help us to understand if the amount of participants is enoug to state that our result are confident or we should increase thair number (or fall into dispair and consired that there is no statistically confident difference at all).

From the statistical point of view, our next hypothesis was as follows: the distribution of answers for a test with a parallel text differs from the corresponding control group distribution. We used statistical power with a 95% confidence level to evaluate the number of participants. By definition of statistical power, a hypothesis can be accepted with 95% confidence level if its power is less than 0.05. 


```python
order = ['Ukranian', 'Belarussian', 'Bulgarian', 
         'Polish', 'Czech', 'Slovak', 'Serbian', 'Slovene']
#powers = pd.DataFrame(columns=["Test "+str(i) for i in range(1, 7)], index=order)
powers = pd.DataFrame(columns=[], index=order)
for i in range(1, 7):
    powers["Test "+str(i)] = -1
powers["All Tests"] = -1
for i in range(1, 7):
    powers["Prtcp "+str(i)] = 0
powers["Total Prtcp"] = -1

for test in range(1, 7):
    data2 = user_data[['parallel_lang', 'mean_mark']][user_data.test_id==test]

    data = data2.groupby(['parallel_lang']).agg(['count', 'mean', 'std'])
    data['se'] = data[('mean_mark', 'std')] / np.sqrt(data[('mean_mark', 'count')])
    data['ci_min'] = data[('mean_mark', 'mean')] - data['se']
    data['ci_max'] = data[('mean_mark', 'mean')] + data['se']
    data.columns = ['count', 'mean', 'std', 'se', 'ci_min', 'ci_max']
    
    for i in data.index:
        powers.loc[i, "Test "+str(test)] = norm.cdf(1.645-abs(data['mean'][i]-data['mean']['No Parallel Text'])/(data['se'][i]))
        powers.loc[i, "Prtcp "+str(test)] = part_no[(part_no['test_id']==test)&(part_no['parallel_lang']==i)]['count'].iloc[0]
        
data2 = user_data[['parallel_lang', 'mean_mark']]

data = data2.groupby(['parallel_lang']).agg(['count', 'mean', 'std'])
data['se'] = data[('mean_mark', 'std')] / np.sqrt(data[('mean_mark', 'count')])
data['ci_min'] = data[('mean_mark', 'mean')] - data['se']
data['ci_max'] = data[('mean_mark', 'mean')] + data['se']
data.columns = ['count', 'mean', 'std', 'se', 'ci_min', 'ci_max']

for i in data.index:
    powers.loc[i, "All Tests"] = norm.cdf(1.645-abs(data['mean'][i]-data['mean']['No Parallel Text'])/(data['se'][i]))
    powers.loc[i, "Total Prtcp"] = part_no[part_no['parallel_lang']==i]['count'].sum()

powers = powers.drop("No Parallel Text")
data_show = powers.style.format('{:5.3f}')

data_show
```

## CONCLUSION

In this paper, we present a study of written intelligibility of eight unknown Slavic languages from West, South and East Slavic subgroups for Russian native speakers. In our experiment modelling word comprehension in a text we used cloze tests made of six pieces of fiction text in Russian. These texts were given with parallel translations into the eight languages of study to over 2200 native Russian participants altogether. 

Results of our experiments confirmed our first hypothesis that a parallel text in an unknown Slavic language increases the amount of correctly restored words in a gapped text in Russian. According to the obtained data the languages of the study can be ranked as follows from most to least intelligible for native speakers of Russian: Ukrainian, Belarusian, Bulgarian, Serbian, Slovene, Polish, Slovak, Czech. 

Experiment results also showed that some pieces of text seemed to be more intelligible than the others (with two of the six tests showing higher intelligibility than the other four, with or without a parallel translation in any language). This led us to the conclusion that, strictly speaking, in such studies we are dealing with the intelligibility of texts rather than the intelligibility of a language as a whole. This should be taken into account when doing intelligibility research.

The second hypothesis was that the intelligibility of a text in a foreign but related language correlates not only with the percentage of cognates, but also with the percentage of cognate synonyms and false friends. Analysing the results of our experiment we found that the average correlation with the percentage of same root words (roughly corresponding to cognates) is as high as 0.74, while the average correlation with cognate synonyms, non-cognates and false friends is negative in most cases. However, though the percentage of cognates seems to define text intelligibility to a large extent, neither this, nor the information on the distribution of words of other classes involved in this study is enough to account for the observed changes in the test intelligibility.

It would be thus reasonable to conclude that the intelligibility of a text in one language to a speaker of a foreign but related language depends primarily on the percentage of cognates (in these two languages in general and in this text in particular) and on how well these can be recognized in this text, but these are not the only defining factors. Further research is required to study the impact of different classes of words on text intelligibility and identify and assess the influence of other possible contributors, such as syntactic differences within the language pair or text features which enhance its intelligibility.
